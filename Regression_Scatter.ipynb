```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_ad_expenditure = pd.read_csv("/Users/a0000/Desktop/Data Vis/Regression Plot/scatter_plot_ii.csv")
```

```python
df_ad_expenditure
```

```python
#第一種方法：用sns.regplot寫出回歸線圖表
sns.set(rc = {'figure.figsize':(9,6)})
sns.regplot (x = "Budget",
            y = "Sales",
            data = df_ad_expenditure,
            #散佈點的顏色設定
            scatter_kws = {'color' : 'k'},
             #迴歸線顏色設定
            line_kws = {'color' : 'red'})

plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales",fontsize = 14, weight = "bold")

plt.show()
```

```python
#第二種方法：用sns.lmplot寫出回歸圖
sns.lmplot(x = "Budget",
           y = "Sales",
          data = df_ad_expenditure,
           #將圖片的長寬設定為10
          height = 10,
          #散佈點的顏色設定
          scatter_kws = {'color' : 'k'},
          #迴歸線顏色設定
          line_kws = {'color' : 'red'})

plt.xlabel("Ad Expenditure in (000's $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales",fontsize = 14, weight = "bold")
plt.show()
```
