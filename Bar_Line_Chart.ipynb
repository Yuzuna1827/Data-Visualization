```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv("/Users/a0000/Desktop/Data Vis/Bar and Line Chart/bar_line_chart_data.csv")
```

```python
df_kdnuggets
```

```python
#這行程式碼創建了一個新的圖形物件（fig）和一個軸物件（ax）。圖形物件用於容納整個圖表，而軸物件用於在圖表上放置資料、標籤和其他元素。
#plt.subplots(): 這是matplotlib中的一個函數，用於創建一個新的圖表和軸物件
fig, ax = plt.subplots(figsize = (10,7))

#這是一個在指定軸物件（ax）上創建條形圖的函數。條形圖通常用於表示不同類別或項目之間的數量或比較關係
ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],
       color = "k")

#將Y軸的標籤設定為"Number of Participants"，同時使用了"weight"參數，將標籤的字體設定為粗體（"bold"
ax.set_ylabel("Number of Participants",
               weight = "bold")
#設定了Y軸的刻度
ax.tick_params(axis = "y",
              width = 2,
              labelsize = "large")

# 這行程式碼創建了一個新的軸物件ax1，它是基於現有軸物件ax的"孿生"軸。這意味著ax1將共享相同的X軸，但具有獨立的Y軸
ax1 = ax.twinx()
#這個函數設置了第二個Y軸 (ax1) 的數值範圍
ax1.set_ylim(0,1)
#格式化第二個Y軸的標籤，以顯示百分比,這是通過PercentFormatter物件來實現的
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))

#ax1.plot(): 這是一個在ax1上繪製線圖的函數
ax1.plot(df_kdnuggets["Year"],
         df_kdnuggets["Python Users"],
         color = "#B60000",
         marker = "D")

ax1.set_ylabel("Python Users",
               color = "#b60000",
              weight = "bold")
ax1.tick_params(axis = "y",
                colors = "#b60000",
               width = 2,
               labelsize = "large")

ax.set_title("KD Nuggets Survey Pyhton Users (2012 - 2019)", fontsize = "14", weight = "bold")
```
