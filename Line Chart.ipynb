```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_spx_ftse_00_10 = pd.read_csv("/Users/a0000/Desktop/Data Vis/Line Chart/line_chart_data.csv")
```

```python
df_spx_ftse_00_10
```

```python
df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"])
```

```python
df_spx_ftse_00_10["new_date"]
```

```python
labels = ["S&P 500","FTES 100"]
plt.figure(figsize = (20,8))
plt.plot(df_spx_ftse_00_10["new_date"],df_spx_ftse_00_10["GSPC500"])
plt.plot(df_spx_ftse_00_10["new_date"],df_spx_ftse_00_10["FTSE100"])
plt.title("S&P vs FTES Returns (2000 - 2010)",fontsize = 14,fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```
